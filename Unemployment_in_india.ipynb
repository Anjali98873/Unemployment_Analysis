```python
import datetime as dt
import calendar
```

```python
# Assuming you have a DataFrame df with a 'Date' column and a 'month_int' column (as in your code)
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['month_int'] = df['Date'].dt.month
df['month_int'].fillna(0, inplace=True)
df['month_int'] = df['month_int'].astype(int)  # Convert to integers

df['month'] = df['month_int'].apply(lambda x: calendar.month_abbr[x])
```

```python
# Assuming you have a DataFrame df_11_2020 with a 'Date' column
df_11_2020['Date'] = pd.to_datetime(df_11_2020['Date'], dayfirst=True)
df_11_2020['month_int'] = df_11_2020['Date'].dt.month
df_11_2020['month'] = df_11_2020['month_int'].apply(lambda x: calendar.month_abbr[x])
```

```python
#visualize labor partipication rate &unemployment rate in each month
df_EEE = df.groupby(['month'])[['Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate']].mean()
df_EEE =pd.DataFrame(df_EEE).reset_index()
month = df_EEE.month
unemployment_rate = df_EEE['Estimated Unemployment Rate']
labor_participation_rate =df_EEE['Estimated Unemployment Rate']

fig = go.Figure()

fig.add_trace(go.Bar(x = month, y = unemployment_rate, name = 'unemployment_Rate'))
fig.add_trace(go.Bar(x = month, y = labor_participation_rate, name = 'Labor Participation Rate'))

fig.update_layout(title ='Unemploment Rate and Labour Participation Rate', xaxis={'categoryorder':'array','categoryarray':['Jan','Feb','MAr','Apr','May','Jun','jul','Aug','Sep','Oct','Nov','Dec']})
fig.show()
```

```python
#visualize labor partipication rate &unemployment rate in each month
df_EEE_11_2020 = df_11_2020.groupby(['month'])[['Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate']].mean()
df_EEE_11_2020 =pd.DataFrame(df_EEE_11_2020).reset_index()
month = df_EEE_11_2020.month
unemployment_rate = df_EEE_11_2020['Estimated Unemployment Rate']
labor_participation_rate =df_EEE_11_2020['Estimated Unemployment Rate']

fig = go.Figure()

fig.add_trace(go.Bar(x = month, y = unemployment_rate, name = 'unemployment_Rate'))
fig.add_trace(go.Bar(x = month, y = labor_participation_rate, name = 'Labor Participation Rate'))

fig.update_layout(title ='Unemploment Rate and Labour Participation Rate', xaxis={'categoryorder':'array','categoryarray':['Jan','Feb','MAr','Apr','May','Jun','jul','Aug','Sep','Oct','Nov','Dec']})
fig.show()
```
